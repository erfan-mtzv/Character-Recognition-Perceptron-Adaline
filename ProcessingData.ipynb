```python
import numpy as np
import pandas as pd
import os
```

# Train dataset

```python
txtFiles = os.listdir('G:\Academic\Soft Computing\Characters-TrainSet')
sList=[]
lables=[]
pixles=[]

for j in txtFiles:
    df = pd.read_csv('G:\Academic\Soft Computing\Characters-TrainSet\{}'.format(j), header=None, names=None, index_col=None)
    A = df.values
    for i in range(A.shape[0]):
        sList.append([*A.tolist()[i][0]])
    pixles.append(np.array(sList).flatten().reshape(-1,63).tolist()[0])
    sList.clear()
    lables.append(list(j)[0])

DF = pd.DataFrame(data = pixles, columns=['pixle ' + str(i) for i in range(1,64)])
DF['lables'] = lables
DF
```

# Replace '.' & '#' with -1 & 1

```python
DF = DF.replace(to_replace = '#', value = 1)
DF = DF.replace(to_replace = '.', value = -1)
DF
```

# Test dataset

```python
txtFiles = os.listdir('G:\Academic\Soft Computing\Characters-TestSet')
sList=[]
lables=[]
pixles=[]

for j in txtFiles:
    df1 = pd.read_csv('G:\Academic\Soft Computing\Characters-TestSet\{}'.format(j), header=None, names=None, index_col=None)
    A = df1.values
    for i in range(A.shape[0]):
        sList.append([*A.tolist()[i][0]])
    pixles.append(np.array(sList).flatten().reshape(-1,63).tolist()[0])
    sList.clear()
    lables.append(list(j)[0])

DF1 = pd.DataFrame(data = pixles, columns=['pixle ' + str(i) for i in range(1,64)])
DF1['lables'] = lables
DF1
```
